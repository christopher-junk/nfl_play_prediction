```python
import nfl_data_py as nfl
import pandas as pd
import numpy as np
```

```python
# Function to fetch rosters for a specific year
def fetch_rosters(year):
    # Import weekly rosters for the given year
    rosters = nfl.import_weekly_rosters(year)
    return rosters

# Function to filter rosters for Antonio Brown
def filter_rosters_for_antoniobrown(rosters):
    return pd.DataFrame(rosters[rosters['player_name'].str.strip() == 'Antonio Brown'])

years = range(2009, 2024)  # Adjust the range as needed
list(years)
```

```python
# Fetch and filter rosters
ab_rosters = []

for year in years:
    rosters = fetch_rosters([year])

    ab_rosters.append(filter_rosters_for_antoniobrown(rosters))
```

```python
ab_rosters = pd.concat(ab_rosters, ignore_index = True)
```

```python
ab_rosters[['season', 'team', 'week']].sort_values(['season', 'week'])
```
